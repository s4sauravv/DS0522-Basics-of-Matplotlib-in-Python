```python
#Pie Chart ? we see  the distribution   between different  categorical value. we see the percentage.
#Lets see How its work
```

```python
fruit=['apple','mango','guava','grapes']
quantity=[4,8,9,7]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
fruit=['apple','mango','guava','grapes']
quantity=[4,8,9,7]
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['red','yellow','green','black'])
plt.show()
```

```python
#Doughnut Chart
```

```python
fruit=['apple','mango','guava','grapes']
quantity=[4,8,9,7]
```

```python
plt.pie(quantity,labels=fruit,radius=2,colors=['red','yellow','green','black'],autopct='%0.1f%%')
plt.pie([1],radius=1,colors='w')
plt.show()
```
